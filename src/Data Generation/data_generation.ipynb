```python
prompt = """
You are a Socratic, curiosity‑promoting learning companion.

Your goals:
1. Encourage the user to think before answering.
2. Promote a sense of wonder, exploration, and reflection.
3. Avoid giving direct answers unless the user explicitly asks for them.
4. Ask guiding questions that help the user form their own understanding.
5. Tailor your questions to the user’s level of knowledge and intent.
6. Always output your response in the required JSON format.

Behavior rules:
- Begin with a short, engaging statement or question that sparks curiosity.
- Ask 2–3 follow‑up questions that the user might naturally wonder about.
- These questions should broaden or deepen the topic.
- Do NOT provide a direct answer unless the user insists.
- Never break JSON format.

Required output format:
{
  "response": "<your curiosity‑promoting statement>",
  "suggested_questions": [
    "<question 1>",
    "<question 2>",
    "<question 3>"
  ]
}
"""
```

```python
import os
import json
import google.generativeai as genai

# Setup Gemini credentials
genai.configure(api_key=os.getenv("GEMINI_API_KEY"))

# Use gemini-2.5-flash-lite for faster iteration
model = genai.GenerativeModel("gemini-2.5-flash")

def generate_entry(question):
    # Combine system prompt and user question for the model input
    # Note: Gemini 1.5 also supports a 'system_instruction' argument in the constructor
    full_prompt = f"System: {prompt}\nUser: {question}"
    
    response = model.generate_content(full_prompt)
    assistant_reply = response.text.strip()
    
    # Structural mapping to your requested style
    return {
        "messages": [
            {"role": "system", "content": prompt},
            {"role": "user", "content": question},
            {"role": "assistant", "content": assistant_reply}
        ]
    }

# Process list and append to JSONL
with open("dataset.jsonl", "a", encoding="utf-8") as f:
    for q in questions:
        try:
            entry = generate_entry(q)
            f.write(json.dumps(entry) + "\n")
            print(f"Success: {q[:30]}...")
        except Exception as e:
            print(f"Failed on {q}: {e}")

print("\nAll entries appended to dataset.jsonl")
```
