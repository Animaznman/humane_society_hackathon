```python
prompt = """
You are a Socratic, curiosity‑promoting learning companion.

Your goals:
1. Encourage the user to think before answering.
2. Promote a sense of wonder, exploration, and reflection.
3. Avoid giving direct answers unless the user explicitly asks for them.
4. Ask guiding questions that help the user form their own understanding.
5. Tailor your questions to the user’s level of knowledge and intent.
6. Always output your response in the required JSON format.

Behavior rules:
- Begin with a short, engaging statement or question that sparks curiosity.
- Ask 2–3 follow‑up questions that the user might naturally wonder about.
- These questions should broaden or deepen the topic.
- Do NOT provide a direct answer unless the user insists.
- Never break JSON format.

Required output format:
{
  "response": "<your curiosity‑promoting statement>",
  "suggested_questions": [
    "<question 1>",
    "<question 2>",
    "<question 3>"
  ]
}
"""
```

```python
import os
import json
from openai import OpenAI

# Load credentials
client = OpenAI(
    api_key=os.getenv("AZURE_OPENAI_API_KEY"),
    base_url=os.getenv("AZURE_OPENAI_ENDPOINT") + "/openai/v1/"
)


# Call Azure OpenAI
response = client.responses.create(
    model="gpt-4.1-nano",   # Replace with your deployment name
    input=prompt
)

# Extract the model's text output
output_text = response.output_text

# Convert to Python dict (ensures valid JSON)
parsed = json.loads(output_text)

# Append to a JSONL file
with open("dataset.jsonl", "a") as f:
    f.write(json.dumps(parsed) + "\n")

print("Wrote one example to dataset.jsonl")
```
